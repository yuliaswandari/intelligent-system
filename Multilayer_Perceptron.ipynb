Import library math

```
import math
```

Inisialisasi parameter

```
x1 = [1, 0, 1, 0]
x2 = [1, 1, 0, 0]
#w13 = 0.5
#w14 = 0.9
#w23 = 0.4
#w24 = 1
#w35 = -1.2
#w45 = 1.1
w =[0.5, 0.9, 0.4, 1, -1.2, 1.1]
#Theta3 = 0.8
#Theta4 = -0.1
#Theta5 = 0.3
Theta = [0.8, -0.1, 0.3]
deltaw = [0, 0, 0, 0, 0, 0]
deltaT = [0, 0, 0]
ytrue = [0, 1, 1, 0]
alpha = 0.1
beta = 0.95
Y = [0, 0, 0, 0]
e = [0, 0, 0, 0]
```

Fungsi sigmoid

```
def sigmoid(name):
  z = 1/(1+math.exp(-(name)))
  return z 
```

Iterasi per data - satu epoch

```
def iterasi():
  for x in range(4):
    y3 = sigmoid(x1[x]*w[0]+x2[x]*w[2]-Theta[0])
    y4 = sigmoid(x1[x]*w[1]+x2[x]*w[2]-Theta[1])
    y5 = sigmoid(y3*w[4]+y4*w[5]-Theta[2])
    Y[x]=y5
    e[x] = ytrue[x]-Y[x]
    delta5 = y5*(1-y5)*e[x]
    delta3 = y3*(1-y3)*delta5*w[4]
    delta4 = y4*(1-y4)*delta5*w[5]
    w[4] = w[4] + alpha*y3*delta5 + beta*deltaw[4]
    w[5] = w[5] + alpha*y4*delta5 + beta*deltaw[5]
    Theta[2] = Theta[2] + alpha*(-1)*delta5 + beta*deltaT[2]
    w[0] = w[0] + alpha*x1[x]*delta3 + beta*deltaw[0]
    w[2] = w[2] + alpha*x2[x]*delta3 + beta*deltaw[2]
    Theta[0] = Theta[0] + alpha*(-1)*delta3 + beta*deltaT[0]
    w[1] = w[1] + alpha*x1[x]*delta4 + beta*deltaw[1]
    w[3] = w[3] + alpha*x2[x]*delta4 + beta*deltaw[3] 
    Theta[1] = Theta[1] +  alpha*(-1)*delta4 + beta*deltaT[1]
    deltaw[4] = alpha*y3*delta5
    deltaw[5] = alpha*y4*delta5
    deltaT[2] = alpha*(-1)*delta5
    deltaw[0] = alpha*x1[x]*delta3
    deltaw[2] = alpha*x2[x]*delta3
    deltaT[0] = alpha*(-1)*delta3 
    deltaw[1] = alpha*x1[x]*delta4
    deltaw[3] = alpha*x2[x]*delta4
    deltaT[1] = alpha*(-1)*delta4
```

Iterasi per epoch

```
for p in range(20000):
    iterasi()
    q = str(p+1)
    print("Hasil epoch ke " + q)
    print("Y = x1 XOR x2")
    print(Y)
    print("Ytrue")
    print(ytrue)
    print ("error")
    print(e)
    q = (e[0]**2+e[1]**2+e[2]**2+e[3]**2)/4
    print("MSE")
    print(q)
    print("bobot")
    print(w)
    print("threshold")
    print(Theta)
    print("------------")
```
