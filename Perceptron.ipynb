Inisialisasi parameter

```
w = [0.3, -0.1]
theta = 0.2
alpha = 0.1
x1 = [0, 0, 1, 1]
x2 = [0, 1, 0, 1]
Yd = [0, 0, 0, 1]
Y = [0, 0, 0, 0]
e = [0, 0, 0, 0]
```

Step function

```
def step(name):
  if name>=0:
    return 1
  else:
    return 0
```

Iterasi per data - satu epoch

```
def iterasi():
  for x in range(4):
    Ys = x1[x]*w[0]+x2[x]*w[1]-theta
    Ys = step(Ys)
    Y[x] = Ys
    e[x] = Yd[x]-Ys
    w[0]=w[0]+alpha*x1[x]*e[x]
    w[1]=w[1]+alpha*x2[x]*e[x]
```

Iterasi per epoch

```
  for p in range(10):
    iterasi()
    q = str(p+1)
    print("Hasil epoch ke " + q)
    print("Y = x1 AND x2")
    print(Y)
    print("Ytrue")
    print(Yd)
    print ("error")
    print(e)
    print("bobot")
    print(w)
    print("------------")
```
